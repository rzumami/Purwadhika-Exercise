### Nomor 6

```python
# Melihat index missing value pada kolom Max.Price

index_mp = data[data['Max.Price'].isna()].index
index_mp
```

```python
# Melihat index missing value pada kolom Fuel.tank.capacity

index_ftc = data[data['Fuel.tank.capacity'].isna()].index
index_ftc
```

### Nomor 7

```python
# Membuat fungsi untuk mengisi nilai missing value

def fillmv(x):
    return x.fillna(x.mean())
```

```python
# Mengisi missing value dengan fungsi yang sudah dibuat

data[['Max.Price','Fuel.tank.capacity']] = data[['Max.Price','Fuel.tank.capacity']].apply(fillmv)
```

### Nomor 8

```python
# Melihat apakah index yang memiliki missing value pada kolom Max.Price sudah terisi atau belum

data.iloc[index_mp]['Max.Price']
```

```python
# # Melihat apakah index yang memiliki missing value pada kolom Fuel.tank.capacity sudah terisi atau belum
data.iloc[index_ftc]['Fuel.tank.capacity']
```

```python
# Membuat sebuah fungsi 
meanprice = data['Price'].mean()

def upper_make(data):
    if data['Price'] > meanprice:
        return data['Make'].upper()
    else:
        return data['Make']
```

```python
data['Upper'] = data[['Price','Make']].apply(upper_make, axis=1)
```
