```python
# Series 1

buah = ['mangga', 'anggur', 'jambu', 'alpukat', 'kelengkeng', 'semangka', 'jeruk', 'jambu', 'markisa', 'pir']
nilai = np.random.uniform(1,50,10)
series1 = pd.Series(data=nilai, index=buah)
series1
```

```python
# Series 2

prov = ['Sumatera', 'Sulawesi', 'Jawa', 'Kepulauan Riau', 'Kalimantan']
kota = ['Aceh', 'Makasar', 'Jakarta', 'Riau', 'Samarinda']
series2 = pd.Series(data=prov, index=kota)
series2
```

```python
# Series 3

random = np.random.randint(10,50,10)
buah_mia = ['anggur', 'mangga', 'apel', 'manggis', 'markisa', 'alpukat', 'aren', 'almon', 'malaka', 'aprikot']
series3 = pd.Series(data=random, index=buah_mia)
series3
```
